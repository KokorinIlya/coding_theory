```python
import numpy as np
import math
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
@np.vectorize
def q(x):
    x_true = 10 ** (x / 10)
    x_arg = math.sqrt(2) * math.sqrt(x_true)
    integral = integrate.quad(lambda t: math.exp(- (t ** 2) / 2), -np.inf, x_arg)[0]
    return 1 - integral / math.sqrt(2 * math.pi)
```

```python
snrs = np.linspace(-2, 12)
ps = q(snrs)
```

```python
plt.semilogy(snrs, ps)
plt.xlabel('Eb/N0 (дБ)')
plt.ylabel('p')
plt.grid()
plt.title('Зависимость вероятности ошибке в единственном бите, без использования кодирования, от отношения сигнал/шум')
plt.show()
```
