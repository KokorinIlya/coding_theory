```python
import math
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def calculate_graismer(k, d):
    total_sum = 0
    for i in range(k):
        total_sum += math.ceil(d / (2 ** i))
    return total_sum
```

```python
k_to_n = {
    3: 14,
    4: 15,
    5: 16,
    6: 18,
    7: 19,
    8: 20,
    9: 21,
    10: 22,
    11: 23,
    12: 24,
    13: 27,
    14: 28,
    15: 30,
    16: 31,
    17: 32,
    18: 33,
    19: 34,
    20: 35,
    21: 36,
    22: 37,
    23: 38,
    24: 39,
    25: 40
}
```

```python
ks = []
ns = []
n_grs = []
for k, n in k_to_n.items():
    n_gr = calculate_graismer(k, d=8)
    ks.append(k)
    ns.append(n)
    n_grs.append(n_gr)
```

```python
plt.figure(dpi=600)
plt.plot(ks, ns, label='n')
plt.plot(ks, n_grs, label='N(k, d)')
plt.legend(loc='upper left')
plt.title('Сопоставление оптимальной длины кода,\
вычисленной по таблице 3.2, и N(k, d), вычисленной по формуле Грайсмера')
plt.show()
```
